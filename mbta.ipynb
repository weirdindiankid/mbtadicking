```python
%matplotlib inline
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = (15, 15)
```

```python
boston = gpd.read_file("Parcels_2018.geojson")
unis = gpd.read_file("Colleges_and_Universities.geojson")
```

```python
boston_base = boston.plot(color='white', edgecolor='black')
boston_base.set_xlim([-71.20, -71.00])
unis.plot(ax=boston_base, marker='o', color='red', markersize=25);
```
